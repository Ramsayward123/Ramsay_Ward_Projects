### **Question Three:** (3 points) Which carrier has the longest cumulative departure delay?

```python
#TODO: Your code goes here
# Group by carrier and sum the departure delays, then sort in descending order
# Get the carrier with the longest delays in sum
group = df.groupby('carrier')['dep_delay'].sum()
max_car = group.idxmax()

print(group.sort_values())
#print(max_car)
#clearly EV has the longest cummulative departure delay
print(f'The carrier with the longest cummulative depatrure delay is: {max_car}')






```
