```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# RK 4 Method
def rk4 (f, x0, t0, tmax, dt):
    x0 = np.array(x0)
    N = int(np.floor((tmax - t0) / dt) + 1)
    t = np.linspace(t0, tmax, N + 1)
    m = len(t)
    k = len(x0)
    x = np.zeros((m, k))
    x[0] = x0
    for n in range(len(t) - 1):
        k1 = f(t[n], x[n])
        k2 = f(t[n] + dt/2, x[n] + dt/2 * k1)
        k3 = f(t[n] + dt/2, x[n] + dt/2 * k2)
        k4 = f(t[n] + dt, x[n] + dt * k3)
        x[n+1] = x[n] + dt * (k1 + 2*k2 + 2*k3 + k4)/6
    return t, x
```

```python
def f(t, x):
    return np.array([x[1], -x[1] - np.sin(x[0])])
```

```python
x0 = [np.pi/6, 0]
t0 = 0
tmax = 15
dt = 0.42
t, x = rk4(f, x0, t0, tmax, dt)
plt.plot(t, x[:, 0])
plt.grid()
plt.show()
```

```python
t0 = 0
tmax = 15
dt = 0.42

fig_ang, ax_ang = plt.subplots(4, 4)
fig_vel, ax_vel = plt.subplots(4, 4)
fig_av, ax_av = plt.subplots(4, 4)

ang0 = [np.pi/6, np.pi/4, np.pi/3, 5/12 * np.pi]
vel0 = [0, 1, 2, 3]

for n, ang in enumerate(ang0):
    for m, vel in enumerate(vel0):
        x0 = [ang, vel]
        t, x = rk4(f, x0, t0, tmax, dt)
        aa = ax_ang[n, m]
        av = ax_vel[n, m]
        aav = ax_av[n, m]
        
        aa.plot(t, x[:, 0])
        av.plot(t, x[:, 1])
        aav.plot(x[:, 0], x[:, 1])
        
        aa.grid()
        av.grid()
        aav.grid()

fig_ang.show()
fig_vel.show()
fig_av.show()
        
```

```python
t0 = 0
tmax = 15
dt = 0.42

fig_ang, ax_ang = plt.subplots(2, 2)
fig_vel, ax_vel = plt.subplots(2, 2)
fig_av, ax_av = plt.subplots(2, 2)

ang0 = [np.pi/6, np.pi/4, np.pi/3, 5/12 * np.pi]
vel0 = [0, 1, 2, 3]

for n, ang in enumerate(ang0):
    aa = ax_ang[n//2, n%2]
    av = ax_vel[n//2, n%2]
    aav = ax_av[n//2, n%2]
    for m, vel in enumerate(vel0):
        x0 = [ang, vel]
        t, x = rk4(f, x0, t0, tmax, dt)

        label = f"w0={vel}rad/s"
        aa.plot(t, x[:, 0], label=label)
        av.plot(t, x[:, 1], label=label)
        aav.plot(x[:, 0], x[:, 1], label=label)

    title = f"x0={ang*180/np.pi:0.0f}deg"
    aa.set_title(title)
    aa.set_xlabel("t")
    aa.set_ylabel("x(t)")
    aa.legend()
    aa.grid()

    av.set_title(title)
    av.set_xlabel("t")
    av.set_ylabel("w(t)")
    av.legend()
    av.grid()

    aav.set_title(title)
    aav.set_xlabel("x(t)")
    aav.set_ylabel("w(t)")
    aav.legend()
    aav.grid()


fig_ang.tight_layout()
fig_ang.show()
fig_ang.savefig("ang_t.png")

fig_vel.tight_layout()
fig_vel.show()

fig_av.tight_layout()
fig_av.show()
fig_av.savefig("vel_ang.png")
```
